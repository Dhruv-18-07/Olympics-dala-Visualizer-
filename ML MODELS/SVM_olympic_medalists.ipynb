```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split 
```

```python
athlete_events = pd.read_csv('../CSV for ML models/athlete_events.csv')
athlete_events.head()
```

```python
filter_data = athlete_events[["Sex", "Age", "Height", "Weight", "Team", "Year", "Season", "Sport", "Event", "Medal"]]
```

```python
# winter_data = filter_data[filter_data["Season"] == "Winter"]
winter_data = athlete_events[athlete_events["Season"] == "Winter"]
winter_data.head()
```

```python
#summer_data = filter_data[(filter_data["Season"] == "Summer")]
summer_data = athlete_events[athlete_events["Season"] == "Summer"]
summer_data.head()
```

```python
summer_sports = summer_data["Sport"].unique()
winter_sports = winter_data['Sport'].unique()
```

```python
df1 = summer_data[(summer_data["Sport"] == "Gymnastics") & (summer_data["Sex"] == "M")]
df1 = df1[["Age", "Height", "Weight", "Medal"]]
df1 = df1.dropna()
df1 = df1.reset_index(drop = True)
df1.head()
```

```python

target = df1["Medal"]
target_names = ["Gold", "Silver", "Bronze"]
```

```python
data = df1.drop("Medal", axis=1)
feature_names = data.columns
data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2,random_state=42)
```

```python
model = SVC(kernel='rbf')
model.fit(X_train, y_train)
```

```python
print('Test Acc: %.3f' % model.score(X_test, y_test))
```

```python
 
# from sklearn.metrics import classification_report
# predictions = model.predict(X_test)
# print(classification_report(y_test, predictions,
#                             target_names=target_names))
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [0.01, 0.1, 1, 10],
    'kernel': ['rbf']
}

grid_search = GridSearchCV(SVC(), param_grid, cv=5)
grid_search.fit(X_train, y_train)

print("Best parameters:", grid_search.best_params_)
model = grid_search.best_estimator_
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```
